```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv(r'D:\projects\task 1\archive\netflix_titles.csv')
```

```python
#display settings

sns.set(style='whitegrid')

plt.rcParams['figure.figsize'] = (10, 6)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['rating'].fillna('Not Rated', inplace=True)
df['country'].fillna('Unknown', inplace=True)
```

```python
sns.countplot(x='type', data=df, palette='Set2')
plt.title('Distribution of Movies and TV Shows on Netflix')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(y='rating', data=df, order=df['rating'].value_counts().index[:10], palette='coolwarm')
plt.title('Top Ratings on Netflix')
plt.show()
```

```python
plt.figure(figsize=(10,6))
df['release_year'].value_counts().sort_index().plot(kind='line')
plt.title('Number of Releases per Year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.show()
```

```python
top_countries = df['country'].value_counts().head(10)
plt.figure(figsize=(8,5))
sns.barplot(x=top_countries.values, y=top_countries.index, palette='viridis')
plt.title('Top 10 Countries Producing Netflix Content')
plt.xlabel('Number of Titles')
plt.show()
```

```python
from collections import Counter

genre_list = df['listed_in'].dropna().apply(lambda x: x.split(', '))
genres = Counter([g for sublist in genre_list for g in sublist])
top_genres = dict(genres.most_common(10))

plt.figure(figsize=(10,6))
sns.barplot(x=list(top_genres.values()), y=list(top_genres.keys()), palette='magma')
plt.title('Top 10 Most Common Genres on Netflix')
plt.xlabel('Count')
plt.show()
```

# Insights:

Movies make up the majority of Netflix content.

The most common ratings are TV-MA and TV-14.

Netflix content production increased significantly after 2015.

The United States produces the most Netflix titles.
